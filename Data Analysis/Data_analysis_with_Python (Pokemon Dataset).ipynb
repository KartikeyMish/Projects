# Question 19. Compare the attacks of Pokemon with highest HP and lowest HP

```python
x=df.nlargest(1, ['HP']) # Data of Pokemon with Highest HP 
y=df.nsmallest(1, ['HP']) # Data of Pokemon with Lowest HP 
# Extrating both the data to new dataframes
E=x.loc[:, ["Name","HP","Attack"]] 
F=y.loc[:, ["Name","HP","Attack"]]
```

```python
E # Highest HP Pokemon Data of Attack
```

```python
F # Lowest HP Pokemon Data of Attack
```

```python
# Combining the data of both the dataframe. 
G= (E,F)
H= pd.concat(G)
H
```

```python
x_axis = np.arange(len(H.Name))

# Multiple colors of bars

plt.bar(x_axis -0.2, H.HP, width=0.4, label = 'HP')
plt.bar(x_axis +0.2, H.Attack, width=0.4, label = 'Attack')

# Xticks :for Stick the X-Axis
plt.xticks(x_axis, H.Name)

#title
plt.title("Comparision between Pokemon Attack with Highest HP and Lowest HP Pokemon")

# Add legend
plt.legend()

# Displaying the graph
plt.show()
```

# Question 20. Compare the defence of Pokemon with Highest HP and lowest HP

```python
A=df.nlargest(1, ['HP']) # Data of Pokemon with Highest HP
B=df.nsmallest(1, ['HP']) # Data of Pokemon with Highest HP
# Extrating both the data to new dataframes
A=A[["Name","HP","Defense"]]  
B=B[["Name","HP","Defense"]]
```

```python
A  # Highest HP Pokemon Data of Defense
```

```python
B # Lowest HP Pokemon Data of Defense
```

```python
# Combining the data of both the dataframe. 
C= (A,B)
D= pd.concat(C)
D
```

```python
x_axis = np.arange(len(D.Name))

# Multiple colors of bars

plt.bar(x_axis -0.2, D.HP, width=0.4, label = 'HP')

plt.bar(x_axis +0.2, D.Defense, width=0.4, label = 'Defense')

# Xticks :for Stick the X-Axis
plt.xticks(x_axis, D.Name)

plt.title("Comaprision Between Defence of Pokemon with Highest HP & Lowest HP Pokemon")

# Add legend
plt.legend()

# Displaying the graph
plt.show()
```
